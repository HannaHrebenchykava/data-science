# Перевірка на нормальний розподіл

<font color='blue'><font size = 3 ><b>У кожному випадку у нас буде:

<font color='blue'><font size = 3 >Нульова гіпотеза <b>(H_0)</b> - розподіл є нормальний.

<font color='blue'><font size = 3 >Альтернативна <b>(H_1)</b> -- розподіл не є нормальним.
    
<font color='blue'><font size = 3 > alpha = 0.05

# Iris-setosa

```python
import scipy

scipy.stats.shapiro(df_setosa['sepal_width'])
```

```python
scipy.stats.shapiro(df_setosa['sepal_length'])
```

```python
scipy.stats.shapiro(df_setosa['petal_length'])
```

```python
scipy.stats.shapiro(df_setosa['petal_width'])
```

```python
df_setosa['sepal_width'].plot.hist(bins=50)
```


<font size = 3 ><mark>Хоча графік розподілу sepal_width візуально близький до нормального розподілу, критерій Шапіро говорить протилежнею 

<font size = 3 ><mark>У всіх випадках з iris-setosa pi value дуже мале (<0.05).  Тому ми відхиляємо Н_0 і приймаємо H_1



# Iris-versicolor

```python
scipy.stats.shapiro(df_versicolor['sepal_width'])
```

<font size = 3 ><mark>pi value <0.05.  Тому ми відхиляємо Н_0 і приймаємо H_1

```python
scipy.stats.shapiro(df_versicolor['sepal_length'])
```

<font size = 3 ><mark>pi value <0.05.  Тому ми відхиляємо Н_0 і приймаємо H_1

```python
scipy.stats.shapiro(df_versicolor['petal_width'])
```

<font size = 3 ><mark>pi value <0.05.  Тому ми відхиляємо Н_0 і приймаємо H_1

```python
scipy.stats.shapiro(df_versicolor['petal_length'])
```

<font size = 3 ><mark>pi value суттєво більше ніж 0.05, тому ми <b>HE</b> можемо відхилити H_0. 
Давайте подивимося на графік:

```python
df_versicolor['petal_length'].plot.hist(bins=50)
```

<font size = 3 ><mark>Також побудуємо графік Q-Q. Для порівняння розподілу virsicolor['petal_length'] з нормальним розподілом

```python
scipy.stats.probplot(df_versicolor['petal_length'], dist="norm", plot=plt)
plt.show()
```

<font size = 3 ><mark> На графіку видно, що невелике відхилення від нормального розподілу відбувається тільки на кінцях. Якщо розглядати розподіл versicolor ['petal_length'] у квартилях 25% - 75%.<b> Він буде нормальним розподілом.
